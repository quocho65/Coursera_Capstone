```python
# Extract a table from web
# thank you to Syed Sadat Nazrul for https://towardsdatascience.com/web-scraping-html-tables-with-python-c9baba21059
import requests
import lxml.html as lh
import pandas as pd
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
#Create a handle, page, to handle the contents of the website
page = requests.get(url)
#Store the contents of the website under doc
doc = lh.fromstring(page.content)
#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
#Create empty list

col=[]
i=0
#For each row, store each first element (header) and an empty list
for t in tr_elements[0]:
    i+=1
    name=t.text_content().replace('\n','')
    print( '%d:"%s"'%(i,name))
    col.append((name,[]))
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(1,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 10, the //tr data is not from our table 
    if len(T)!=3:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content().replace('\n','') 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
```

```python
df.drop(df[df['Borough'] == 'Not assigned'].index, inplace = True) 
```

```python
# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library
```

```python
from geopy.geocoders import Nominatim
address = 'Toronto'

geolocator = Nominatim(user_agent='toronto_explorer')
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values

map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df_full['Latitude'], df_full['Longitude'], df_full['Borough'], df_full['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

```python
toronto_data = df_full
toronto_data.head()
```

```python
toronto_data.set_index(['Postal Code','Borough','Latitude','Longitude'])
toronto_neighborhood = toronto_data.apply(lambda x: x.astype(str).str.split(' / ').explode()) 
toronto_neighborhood.set_index('Neighborhood')
```

```python
toronto_neighborhood_onehot = pd.get_dummies(toronto_neighborhood,columns=['Postal Code','Borough'])
```

```python
toronto_neighborhood_onehot.drop('Neighborhood',axis=1,inplace=True)
```

```python
toronto_neighborhood_onehot
```

```python
toronto_neighborhood_onehot.dtypes
```

```python
from sklearn.cluster import KMeans 
# set number of clusters
kclusters = 5

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_neighborhood_onehot)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
toronto_neighborhood_onehot = toronto_neighborhood.assign(cluster=kmeans.labels_)
```

```python
toronto_neighborhood_onehot.head()
```

```python
toronto_neighborhood_onehot.dtypes
```

```python
toronto_neighborhood_onehot.cluster = toronto_neighborhood_onehot.cluster.astype(int)
toronto_neighborhood_onehot.Latitude = toronto_neighborhood_onehot.Latitude.astype(float)
toronto_neighborhood_onehot.Longitude = toronto_neighborhood_onehot.Longitude.astype(float)
```

```python
toronto_neighborhood_onehot.dtypes
```

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as colors
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_neighborhood_onehot['Latitude'], toronto_neighborhood_onehot['Longitude'], toronto_neighborhood_onehot['Neighborhood'], toronto_neighborhood_onehot['cluster']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
